## Handling Outliers

```python
sns.boxplot(data=gfm)
```

```python
sns.boxplot(x='Number_of_Donors', data=gfm)
```

```python
Q1 = np.percentile(gfm['Number_of_Donors'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(gfm['Number_of_Donors'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
  
print("Old Shape: ", gfm.shape)
  
# Upper bound
upper = np.where(gfm['Number_of_Donors'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(gfm['Number_of_Donors'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
gfm.drop(upper[0], inplace = True)
gfm.drop(lower[0], inplace = True)
  
print("New Shape: ", gfm.shape)
  
sns.boxplot(x='Number_of_Donors', data=gfm)
```
