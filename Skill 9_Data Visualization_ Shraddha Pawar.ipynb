## Data Visualization

### Univariate analysis

```python
plt.plot("Sales",data=df,marker = "*")

font1 = {'family':'serif','color':'green','size':20}
plt.title("Line plot",fontdict = font1)
plt.show()
```

The above plot shows the sales of the suprestore.

Maximum sales is upto 5000.

```python
plt.plot("Profit",data=df,marker = "o",color="b", linestyle = "dotted")
font1 = {'family':'serif','color':'green','size':20}
plt.title("Line plot",fontdict = font1)
plt.show()
```

The above plot shows the profit and loss of superstore.

```python
# Countplot

sns.countplot(x='Quantity', data=df )
plt.title("Count plot")
```

Maximum 2 to 3 products purchased in superstore.

### Bivariate analysis

```python
sns.set_context('paper')
 
# create plot
sns.barplot(x = 'Profit/Loss', y = 'Profit', data = df,
            palette = 'PuRd')
font1 = {'family':'serif','color':'black','size':20}
plt.title("Profit loss Plot",fontdict = font1)
plt.legend()
plt.show()
```

From the above plot, loss of superstore is maximum than profit.

```python
fig = px.pie(df, values='Sales', names='Category')
fig.show()
```

```python
# Scatter plot

sns.scatterplot(x='Profit', y='Sales', hue='Profit/Loss', data=df, palette ='YlOrRd' )
font1 = {'family':'serif','color':'black','size':20}
plt.title("Superstore Dataset",fontdict = font1)
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
sns.set_context('paper')
 
# create plot
sns.countplot(x = 'Segment', hue = 'Profit/Loss' ,data = df, palette = 'magma')
plt.show()
```

```python
sns.set_context('paper')
 
# create plot
sns.barplot(x = 'Quantity', y = 'Sales', data = df,
            palette = 'PuRd')
font1 = {'family':'serif','color':'black','size':20}
plt.title("Sales Plot",fontdict = font1)
plt.legend()
plt.show()
```

### Multivariate Analysis

```python
sns.pairplot(data = df,hue = 'Profit/Loss',diag_kind = "kde",kind = "scatter",palette = "husl")
```

```python
# Correlation plot

corr=df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlOrRd")
```
